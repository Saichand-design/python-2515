```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/diabetic.txt')
df.info()
```

```python
df['dstat'] = np.where(df['Diabetic Status']== 'Yes',1,0)
df
```

```python
inputs = df.drop(['Diabetic Status','dstat'],axis= 1)
inputs = inputs.to_numpy()
inputs.shape
```

```python
def scale(c):
   return (c-c.mean())/c.std()
```

```python
def scaleMatrix(m):
  for i in range(m.shape[1]):
    m[:,i] = scale(m[:,i])
  return m
```

```python
inputs_org = inputs.copy()
```

```python
inputs = scaleMatrix(inputs)
inputs
```

```python
for i in range(inputs.shape[1]):
  print(f"col {i} std {inputs[:,i].std()}")
```

```python
ones = np.ones(df['dstat'].count())
len(ones)
```

```python
# feature Matrix
X = np.c_[ones,inputs]
X.shape
```

```python
# label Matrix
y = df['dstat'].to_numpy()
Y = np.c_[y]
Y.shape
```

```python
# random weight
np.random.seed(101)
W1 = np.random.random((6,8)) * 2-1
W2 = np.random.random((8,10)) * 2- 1
W3 = np.random.random((10,8))*2-1
W4 = np.random.random((8,1))* 2- 1
```

```python
def sigmoid (x,w): #activation function at each layer
  return 1/(1+np.exp(-np.dot(x,w)))
```

```python
def mse(y,ycap):
   return ((y - ycap)**2).mean()
```

```python
def deriavte(p):
  return p*(1-p)
```

```python
#Traing Loop
ploss = 0
flag = 0
conv = 0.0000001
for i in range(100000):
  #forward propagation
  a1 = sigmoid(X,W1)
  a2 = sigmoid(a1,W2)
  a3 = sigmoid(a2,W3)
  a4 = sigmoid(a3,W4)
  e4 = Y - a4
  closs = mse(Y,a4)
  delta4 = e4*deriavte(a4)
  e3 = delta4.dot(W4.T)
  delta3 = e3*deriavte(a3)
  e2 = delta3.dot(W3.T)
  delta2 = e2*deriavte(a2)
  e1 = delta2.dot(W2.T)
  delta1 = e1*deriavte(a1)
  W1 += X.T.dot(delta1) *0.02
  W2 += a1.T.dot(delta2) *0.02
  W3 += a2.T.dot(delta3) *0.02
  W4 += a3.T.dot(delta4) *0.02
  if np.abs(ploss - closs)<= conv:
    print(f"training complete adter{i+1} iteration")
    flag = 1
    break
  ploss = closs
if flag == 0:
  print(" bad training")
```

```python
def predict(X,W): # W is array of [ W1,W2,W3,W4]
 a = X
 for i in range(len(W)):
  a = sigmoid(a,W[i])
  # iteration 1 -> sigmoid (X,W1)
  # iteraion 2 -> sigmoid(a,W2),
  # iteration 3 -> sigmoid(a,W3)
  # iteration 4 -> sigmoid (a,w4)
  return a
```

```python
W = [W1,W2,W3,W4]
```

```python
ycap = predict(X,W)
ycap.shape
```
