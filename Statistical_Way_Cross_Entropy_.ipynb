```python
X = df.drop('Diabetic Status', axis=1).values
Y = df['Diabetic Status'].values.reshape(-1, 1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def cross_entropy(y_true, y_pred):
    # Avoid log(0) by clipping
    y_pred = np.clip(y_pred, 1e-15, 1 - 1e-15)
    return -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))
```

```python
# Initialize weights and bias
np.random.seed(101)
weights = np.random.randn(X_scaled.shape[1], 1)
bias = 0
lr = 0.01
epochs = 10000

for i in range(epochs):
    # Forward pass
    z = np.dot(X_scaled, weights) + bias
    y_pred = sigmoid(z)

    # Compute loss
    loss = cross_entropy(Y, y_pred)

    # Backward pass
    dz = y_pred - Y
    dw = np.dot(X_scaled.T, dz) / len(Y)
    db = np.mean(dz)

    # Update weights
    weights -= lr * dw
    bias -= lr * db

    if i % 1000 == 0:
        print(f"Iteration {i}, Cross-Entropy Loss: {loss:.4f}")
```
