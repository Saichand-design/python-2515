````python
import pandas as pd
import numpy as np
````

````python
df = pd.read_csv("/content/diabetic.txt")
df
````

````python
df['dstat'] = np.where(df['Diabetic Status'] == 'Yes', 1, 0)
df['dstat']
````

````python
xdf = df.drop(['Diabetic Status','dstat'],axis =1)
ones = np.ones(df["Age"].count())
X = np.c_[ones,xdf]
X.shape
````

````python
y= df['dstat'].to_numpy()
y = np.c_[y]
y.shape
````

````python
# statistical way :
def weights(x,y):
  from numpy.linalg import inv
  left = inv(x.T.dot(x))
  rgt = x.T.dot(y)
  return left.dot(rgt)
````

````python
W = weights(X,y)
W
````

````python
# predication
def predict(X,W):
  return 1/(1+np.exp(-X.dot(W)))
````

````python
ycap = predict(X,W)
ycap
````

````python
ycap.shape
````

````python
np.c_[y,ycap]
````

````python
yc = np.where(ycap>0.5,1,0)
yc
````

````python
np.c_[y,yc]
````

````python
np.where(y==yc)[0].size
````

````python
def accuracy(y,ycap):
  pcnt = np.where(y==ycap)[0].size
  n = y.shape[0]
  acc = round(pcnt/n*100,2)
  return acc
````

````python
accuracy(y,yc)
````

````python
from sklearn.metrics import accuracy_score
accuracy_score(y,yc)
````

````python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X,y)
yc = model.predict(X)
acc = accuracy_score(y,yc)
acc
````
