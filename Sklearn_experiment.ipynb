````python
import pandas as pd
import numpy as np
````

````python
df = pd.read_csv('/content/diabetic_scores.txt')
df.info()
````

````python
# step 1 : input matrix(feature matrix)
xdf = df.drop(['Diabetic Status','diabetic_score'],axis=1)
xdf.head()
````

````python
X = xdf.to_numpy()
X.shape
````

````python
df.head()
````

````python
Y = df['diabetic_score'].to_numpy()
Y.shape
````

````python
# model object
model = LinearRegression()
# train model
model.fit(X,Y)
# test accuracy
from sklearn.metrics import r2_score
Ycap = model.predict(X)
Ycap
````

````python
np.c_[Y,Ycap]
````

````python
acc = r2_score(Y,Ycap)*100
acc
````

````python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
model.fit(X,Y)
Ycap = model.predict(X)
acc = r2_score(Y,Ycap)*100
acc
````

````python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=10)
model.fit(X,Y)
Ycap = model.predict(X)
acc = r2_score(Y,Ycap)*100
acc
````

````python
from sklearn.svm import SVR
model = SVR(kernel='linear') # other options --> p
model.fit(X,Y)
Ycap = model.predict(X)
acc = r2_score(Y,Ycap)*100
acc
````

````python
from sklearn.neighbors import KNeighborsRegressor
model = KNeighborsRegressor()
model.fit(X,Y)
Ycap = model.predict(X)
acc = r2_score(Y,Ycap)*100
acc
````

````python
from sklearn.neural_network import MLPRegressor
model = MLPRegressor()
model.fit(X,Y)
Ycap = model.predict(X)
acc = r2_score(Y,Ycap)*100
acc
````

````python
df = pd.read_csv('/content/diabetic.txt')
df.info()
````

````python
df['dstat'] = np.where(df['Diabetic Status']=='Yes',1,0)
df.head()
````

````python
xdf = df.drop(['Diabetic Status','dstat'],axis=1)
xdf.head()
````

````python
X = xdf.to_numpy()
X.shape
````

````python
Y = df['dstat'].to_numpy()
Y
````

````python
Y.shape
````

````python
# experiment with logistic regression
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X,Y)
Ycap = model.predict(X)
from sklearn.metrics import accuracy_score
acc = accuracy_score(Y,Ycap)*100
acc
````

````python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X,Y)
Ycap = model.predict(X)
from sklearn.metrics import accuracy_score
acc = accuracy_score(Y,Ycap)*100
acc
````

````python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X,Y)
Ycap = model.predict(X)
from sklearn.metrics import accuracy_score
acc = accuracy_score(Y,Ycap)*100
acc
````

````python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X,Y)
Ycap = model.predict(X)
from sklearn.metrics import accuracy_score
acc = accuracy_score(Y,Ycap)*100
acc
````

````python
from sklearn.svm import SVC
model = SVC(kernel ='linear')
model.fit(X,Y)
Ycap = model.predict(X)
from sklearn.metrics import accuracy_score
acc = accuracy_score(Y,Ycap)*100
acc
````
