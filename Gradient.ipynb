```python
def scale(X):
    return (X-X.mean())/X.std()
```

```python
X_original = X.copy()
Y_original = Y.copy()
def scalematrix(X):
  for i in range(X.shape[1]):
    if i!=0:
      X[:,i] = scale(X[:,i])
  return X
```

```python
#trail 3
Ycap = X.dot(W)
Loss = ((Y-Ycap)**2).mean()
delta = X.T.dot(Y-Ycap)/Y.size
W+= delta *0.02
print("current loss",Loss)
print("latest weight",W)
```

```python
Convergence = 0.000001
ploss = 0
for i in range(100000):
    Ycap = X.dot(W)
    Closs = ((Y-Ycap)**2).mean()
    if i%10000==0:
        print(f"current loss after {i+1} iteration,{Closs}")
    delta = X.T.dot(Y-Ycap)/Y.size
    W+= delta *0.02
    if np.abs(ploss - Closs)<= Convergence:
        print(f" Training is completed after {i+1} iteration")
        flag = 1
        break
    ploss = Closs
if flag==0:
    print("Training is not completed,run few more iterations")
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/diabetic_scores.txt')
df.info()
```

```python
xdf = df.drop(columns=['Diabetic Status','diabetic_score'],axis =1)
xdf
```

```python
inputs = xdf.to_numpy()
inputs
```

```python
ones = np.ones(df['Age'].count())
ones
```

```python
X = np.c_[ones,inputs]
X.shape
```

```python
outputs = df['diabetic_score'].to_numpy()
outputs
```

```python
Y = np.c_[outputs]
Y.shape
```

```python
from numpy.linalg import inv
W = inv(X.T.dot(X)).dot(X.T.dot(Y))
W
```

```python
def predict(X,W):
  return X.dot(W)
```

```python

def rsquare(Y,Ycap):
  rss = ((Y-Ycap)**2).sum()
  tss = ((Y-Y.mean())**2).sum()
  rs= 1 - rss/tss
  return rs
```

```python
Ycap_stats = predict(X,W)
Ycap_stats
```

```python
rsquare(Y,Ycap_stats)
```

```python
X_scale = scalematrix(X)
X_scale
```

```python
Y_scale = scale(Y)
Y_scale
```

```python
for i in range(X_scale.shape[1]):
  if i!=0:
   print(X_scale[:,i].std())
```

```python
Y_scale.std()
```

```python
#random weights
np.random.seed(101)
r = X.shape[1]
c = Y.shape[1]
W = np.random.random((r,c))*2-1
W
```

```python
def mse(Y,Ycap):
  return ((Y-Ycap)**2).mean()
```

```python
#training
Convergence = 0.0000001
flag =0
ploss = 0
for i in range(100000):
  Ycap = predict(X_scale,W)
  Closs = mse(Y_scale,Ycap)
  if i%1000==0:
    print(f"current loss after {i+1} iteration,{Closs}")
  delta = X_scale.T.dot(Y_scale - Ycap)/Y_scale.size
  W+= delta *0.02
  if np.abs(ploss - Closs)<= Convergence:
    print(f" Training is completed after {i+1} iteration")
    flag = 1
    break
  ploss = Closs
if flag==0:
  print("training not completed")
```
