```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/diabetic.txt')
df.info()
```

```python
df['Diabetic Status']
```

```python
df['dstat'] = np.where(df['Diabetic Status']=='Yes',1,0)
df
```

```python
inputs = df.drop(['Diabetic Status','dstat'],axis= 1)
inputs = inputs.to_numpy()
inputs.shape
```

```python
inputs_original = inputs.copy()
```

```python
def scale(x):
  return (x - x.mean())/x.std()
def scalemetrix(x):
    for i in range(x.shape[1]):
        x[:,i] = scale(x[:,i])
    return x
```

```python
inputs = scalemetrix(inputs)
inputs
```

```python
ones = np.ones(inputs.shape[0])
ones
```

```python
#feature metrix
X = np.column_stack((ones,inputs))
X.shape
```

```python
Y = df['dstat'].to_numpy()
Y = np.c_[Y]
Y.shape
```

```python
#intittiate Random weights
np.random.seed(101)
r = X.shape[1]
c = Y.shape[1]
W = np.random.random((r,c))*2 -1
W.shape
W
```

```python
# note : for classification model, no need to scale label
```

```python
def predict(x,w):
    return 1/(1+np.exp(-np.dot(x,w)))
def mse (y,ycap):
    return ((y - ycap)**2).mean()
def derivate(p): #on crossEntropy
    return p*(1-p)
```

```python
# training Loop
conv = 0.0000001
ploss = 0
flog = 0
for i in range(1000000):
    ycap = predict(X,W)
    e = Y - ycap
    closs = mse(Y,ycap)
    if i%1000 == 0:
      print(f"current loss after {i+1} iterations {closs}")
    delta = e*derivate(ycap)
    W+= X.T.dot(delta)*0.02
    if np.abs(ploss-closs)<= conv:
      print(f"Training completed after {i +1} iterations")
      flag = 1
      break
    ploss = closs

if flag == 0:
  print("Training not completed, run few more iterations")
```

```python
ycap = predict(X,W)
ycap
```

```python
# convert probilities into numeric Labels
yc = np.where(ycap>0.5,1,0)
yc
```

```python
np.c_[Y,yc]
```

```python
pcnt = np.where(Y==yc)[0].size
n = Y.shape[0]
acc = pcnt/n*100
acc
```

```python
def accuracy(y,ycap):
     pcnt = np.where(y==ycap)[0].size
     n = y.shape[0]
     acc = pcnt/n*100
     return acc
```
