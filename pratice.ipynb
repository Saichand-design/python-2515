```python
import numpy as np
import pandas as pd
```

```python
age = [20,30,25,40,27,28,22,24,31,29]
len(age)
```

```python
exp = [0,5,2,10,4,3,5,1,2,6]
len(exp)
```

```python
age = np.array(age)
exp = np.array(exp)
```

```python
income = age**2 + exp *3+5
income
```

```python
np.c = [age,exp,income]
```

```python
ones = np.ones(len(age))
ones
```

```python
X = np.c_[ones,age,exp]
```

```python
X.shape
```

```python
Y = np.c_[income]
```

```python
Y.shape
```

```python
W = np.random.random((3,1))
```

```python
W
```

```python
#trail - 1
```

```python
Ycap = X.dot(W)
```

```python
Ycap
```

```python
#loss
mse = ((Y-Ycap)**2).mean()
```

```python
mse
```

```python
#derivates
delta = X.T.dot(Y-Ycap)/len(age)
```

```python
delta
```

```python
#adjust weight

W+= delta *0.02
print("current loss",mse)
print("latest weight",W)
```
